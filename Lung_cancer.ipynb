```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("/content/drive/MyDrive/Data Analytics/lung cancer survey.csv")
df
```

**How many people are diagnosed with lung cancer?**

```python
# Plotting the distribution of lung cancer patients
plt.figure(figsize=(6, 4))
ax = sns.countplot(x='LUNG_CANCER', data=df)

# Add the count on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.title('Distribution of Lung Cancer Diagnosis')
plt.xlabel('Lung Cancer Diagnosis')
plt.ylabel('Number of People')
plt.show()
```

**How many lung cancer patients are male and how many are female?**

```python
# Filter the dataframe to include only individuals diagnosed with lung cancer
lung_cancer_df = df[df['LUNG_CANCER'] == 'YES']

# Count the number of males and females within the lung cancer group
gender_counts = lung_cancer_df['GENDER'].value_counts().reset_index()
gender_counts.columns = ['Gender', 'Count']

# Plotting the distribution of gender within lung cancer patients
plt.figure(figsize=(6, 4))
ax = sns.barplot(x='Gender', y='Count', data=gender_counts)

# Add the count on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.title('Gender Distribution of Lung Cancer Patients')
plt.xlabel('Gender')
plt.ylabel('Number of Patients')
plt.show()
```

**Among the lung cancer patients, how many of them are smokers and how many are non-smokers?**

```python
# Filter the dataframe to include only individuals diagnosed with lung cancer
lung_cancer_df = df[df['LUNG_CANCER'] == 'YES'].copy()

# Map SMOKING column values to descriptive labels for clarity in plotting
lung_cancer_df['SMOKING_STATUS'] = lung_cancer_df['SMOKING'].apply(lambda x: 'Smokers' if x == 2 else ('Non-smokers' if x == 1 else 'Unknown'))

# Count the number of smokers and non-smokers within the lung cancer group
smoking_counts = lung_cancer_df['SMOKING_STATUS'].value_counts().reset_index()
smoking_counts.columns = ['Smoking Status', 'Count']

# Plotting the distribution of smoking status within lung cancer patients
plt.figure(figsize=(7, 5))
ax = sns.barplot(x='Smoking Status', y='Count', data=smoking_counts)

# Add the count on top of the bars
for p in ax.patches:
    ax.annotate(f'{p.get_height()}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 5), textcoords='offset points')

plt.title('Smoking Status of Lung Cancer Patients')
plt.xlabel('Smoking Status')
plt.ylabel('Number of Patients')
plt.show()
```

**At what age are most people diagnosed with lung cancer?**

```python
# Filter the dataframe to include only individuals diagnosed with lung cancer
lung_cancer_df = df[df['LUNG_CANCER'] == 'YES'].copy()

# Find the age with the highest number of lung cancer diagnoses
most_common_age = lung_cancer_df['AGE'].value_counts().idxmax()
count_at_most_common_age = lung_cancer_df['AGE'].value_counts().max()

print(f"The age at which most people are diagnosed with lung cancer is: {most_common_age}")
print(f"Number of diagnoses at this age: {count_at_most_common_age}")

# Optionally, plot the distribution of ages for lung cancer patients
plt.figure(figsize=(12, 6))
sns.histplot(data=lung_cancer_df, x='AGE', bins=20, kde=True)
plt.title('Distribution of Age for Lung Cancer Patients')
plt.xlabel('Age')
plt.ylabel('Number of Patients')
plt.show()
```

**What is the most common symptom in lung cancer patients?**

```python
# Analyze symptoms for lung cancer patients
lung_cancer_df = df[df['LUNG_CANCER'] == 'YES'].copy()

# **DEBUGGING STEP:** Print the columns of the DataFrame to verify the names
print("Columns in lung_cancer_df:")
print(lung_cancer_df.columns)

# Select symptom columns (assuming columns from 'YELLOW_FINGERS' onwards are symptoms)
# You might need to adjust the column names based on your actual dataset
symptom_columns = ['YELLOW_FINGERS', 'ANXIETY', 'PEER_PRESSURE', 'CHRONIC DISEASE',
                   'FATIGUE', 'ALLERGY', 'WHEEZING', 'ALCOHOL CONSUMING',
                   'COUGHING', 'SHORTNESS OF BREATH', 'SWALLOWING DIFFICULTY',
                   'CHEST PAIN'] # Add all symptom column names here

# Calculate the percentage of lung cancer patients experiencing each symptom
symptom_percentages = {}
total_lung_cancer_patients = len(lung_cancer_df)

for symptom in symptom_columns:
    # Assuming 2 represents the presence of the symptom and 1 the absence
    # You might need to adjust this based on how your data is encoded
    # Added a try-except block to gracefully handle potential KeyErrors for other symptom columns if any
    try:
        symptom_count = lung_cancer_df[symptom].value_counts().get(2, 0) # Get count where value is 2, default to 0 if not found
        symptom_percentages[symptom] = (symptom_count / total_lung_cancer_patients) * 100
    except KeyError:
        print(f"Warning: Symptom column '{symptom}' not found in the DataFrame. Skipping.")


# Convert to a DataFrame for easier plotting
symptom_df = pd.DataFrame(list(symptom_percentages.items()), columns=['Symptom', 'Percentage'])

# Sort by percentage to highlight the most common symptoms
symptom_df = symptom_df.sort_values(by='Percentage', ascending=False)

# Plotting the percentage of lung cancer patients experiencing each symptom
plt.figure(figsize=(14, 8))
ax = sns.barplot(x='Percentage', y='Symptom', data=symptom_df, palette='viridis')

# Add the percentage on the bars
for p in ax.patches:
    width = p.get_width()
    plt.text(width + 1, p.get_y() + p.get_height() / 2.,
             '{:1.1f}%'.format(width),
             ha='left', va='center')

plt.title('Percentage of Lung Cancer Patients Experiencing Specific Symptoms')
plt.xlabel('Percentage (%)')
plt.ylabel('Symptom')
plt.xlim(0, 100) # Set x-axis limit to 100%
plt.tight_layout()
plt.show()

# Print the most common symptoms
print("\nMost common symptoms among people with lung cancer:")
print(symptom_df.head()) # Print the top symptoms
```
